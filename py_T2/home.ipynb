```python
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_absolute_error,mean_squared_error,r2_score
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.preprocessing import PolynomialFeatures
from sklearn.linear_model import LinearRegression
df= pd.read_csv('windpower.csv')
df.dropna(inplace=True)

x=df[['Wind speed (m/s)']]
y=df[['Power (kW)']]
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.25,random_state=42)
mse=[]
r2=[]
for i in range(3,7):
    poly=PolynomialFeatures(degree=i)
    x_train_poly=poly.fit_transform(x_train)
    x_test_poly=poly.transform(x_test)
    model=LinearRegression()
    model.fit(x_train_poly,y_train)
    y_prid= model.predict(x_test_poly)
    print("mse",mean_squared_error(y_test,y_prid))
    print("r2",r2_score(y_test,y_prid))
    mse.append(mean_squared_error(y_test,y_prid))
    r2.append(r2_score(y_test,y_prid))
    print('coef',model.coef_)
    print('intersept',model.intercept_)
    for i in [5,10,15,20]:
        print('prid')
        print(model.predict(poly.transform([[i]])))
        plt.scatter(x_test,y_prid)
plt.scatter(x_test,y_prid)
plt.show()
plt.plot(range(3,7),mse)
plt.show()
plt.plot(range(3,7),r2)
plt.show()
```
