```python
import pandas as pd
df = pd.DataFrame({
    "Student":["A","B","C","D"],
    "Math":[70,45,90,60],
    "Science":[75,40,85,55]
})

def grade(g):
    if g['Average']>=80:
        return 'Distincton'
    elif g['Average']>=60:
        return 'First '
    else:
        return "second"

df2=df
df2['Average']=df.apply(lambda x: (x['Math']+x['Science'])/2,axis=1)
df2['Grade']=df.apply(grade,axis=1)
df2
```

```python
df1 = pd.DataFrame({
    "ID":[1,2,3],
    "Name":["A","B","C"]
})
df2 = pd.DataFrame({
    "ID":[2,3,4],
    "Marks":[80,90,70]
})

df3=pd.concat([df1,df2],join='outer',axis=0)
print(df3.isnull().sum().sum())
print(df3.sort_values(by='ID'))
```

```python
import pandas as pd
df = pd.DataFrame({
    "Dept":["IT","IT","HR","HR","Finance","Finance"],
    "Salary":[50000,70000,40000,60000,80000,75000]
})

ans=df.groupby('Dept').agg({'Salary':['min','median','max']})
ans2=df.groupby('Dept')['Salary'].agg(['min','median','max'])
print(ans)
print(ans2)
```

```python
import pandas as pd
import pandas as pd
df = pd.DataFrame({
    "Product":["P1","P2","P3","P4","P5"],
    "Price":[100,200,150,300,250],
    "Quantity":[5,2,4,1,3]
})

df['TotalAmount']=df.apply(lambda x: x['Price']*x['Quantity'],axis=1)
df=df.sort_values(by='TotalAmount',ascending=False)
```

```python
import pandas as pd
df = pd.DataFrame({
    "Employee":["A","B","C","D","E"],
    "Dept":["IT","HR","IT","Finance","HR"],
    "Salary":[50000,40000,70000,80000,45000]
})

df=df.groupby('Dept')['Salary'].agg(['sum','mean'])
df=df[df['mean']>50000]
df
```
