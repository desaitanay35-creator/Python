```python

df5 = pd.DataFrame({
    "Marks1": [85, 70, 90],
    "Marks2": [80, 65, 95],
    "Marks3": [75, 60, 85]
})
def grade(total):
    if total >= 250:
        return "A"
    elif total >= 200:
        return "B"
    else:
        return "C"
df5["Total"] = df5.apply(lambda x: x.sum(), axis=1)
df5["Grade"] = df5["Total"].apply(grade)
print(df5)
```

```python
import networkx as nx
import matplotlib.pyplot as plt
# 🔹
 # 1. Undirected Graph
G = nx.Graph()
# Add nodes
G.add_node(1)
G.add_nodes_from(['A', 'B', 'C'])
G.add_edge(3, 1)
G.add_edges_from([('A', 'B'), ('A', 'C'), ('B', 'C')])
# Add self-loop
G.add_edge('A', 'A')
print("Undirected Graph Nodes:", G.nodes())
print("Undirected Graph Edges:", G.edges())
# Draw Undirected Graph
plt.figure(figsize=(5,5))
nx.draw(G, with_labels=True, node_size=1000, node_color='red',
        width=3, edge_color='blue')
plt.title("Undirected Graph with Loop")
plt.show()
```
