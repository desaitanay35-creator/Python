```python
import requests ,json
import pandas as pd 
```

```python
API_key='xyz'
city_name= 'Ahmedabad'
url=f"http://api.openweathermap.org/geo/1.0/direct?q={city_name},&appid={API_key}"
response = requests.get(url)
# response.status_code
result= response.json()
result
```

```python
lat=result[0]['lat']
lon=result[0]['lon']
```

```python
lat,lon
```

```python
url2=f"http://api.openweathermap.org/data/2.5/air_pollution?lat={lat}&lon={lon}&appid={API_key}"
response = requests.get(url2)
response.status_code
result2= response.json()
for i,j in result2['list'][0]['components'].items():
    print(i," : ",j)
```

```python
url3=f"http://api.openweathermap.org/data/2.5/air_pollution/forecast?lat={lat}&lon={lon}&appid={API_key}"
response = requests.get(url3)
response.status_code
result= response.json()
result
```

```python
d={'aqi':[],'co':[],'no':[],'no2':[],'o3':[],'so2':[],'pm2_5':[],'pm10':[],'nh3':[]}
for i in result['list']:
    d['aqi'].append(i['main']['aqi'])
    d['co'].append(i['components']['co'])
    d['no'].append(i['components']['no'])
    d['no2'].append(i['components']['no2'])
    d['o3'].append(i['components']['o3'])
    d['so2'].append(i['components']['so2'])
    d['pm2_5'].append(i['components']['pm2_5'])
    d['pm10'].append(i['components']['pm10'])
    d['nh3'].append(i['components']['nh3'])
df= pd.DataFrame(d)
df
```

```python
url4=f'https://api.openweathermap.org/data/2.5/weather?lat={lat}&lon={lon}&appid={API_key}'
response= requests.get(url4)
response.status_code
result=response.json()
result
```

```python
print('temp min',result['main']['temp_min'])
print('temp max',result['main']['temp_max'])
print(result['main']['pressure'])
print(result['weather'][0]['description'])
print(result['wind']['speed'] )
print(result['clouds']['all'] )
print(result['visibility'])
```

```python
url5=f'http://api.openweathermap.org/data/2.5/forecast?lat={lat}&lon={lon}&appid={API_key}'
response = requests.get(url5)
result=response.json()
result

d={'dt_txt':[],'temp':[],'temp_min':[],'temp_max':[],'pressure':[],'description':[],'humidity':[],'wind_speed':[],'clouad_all':[],'visibility':[]}
    
for i in result['list']:
    
     d['description'].append(i['weather'][0]['description'])
     d['dt_txt'].append(i['dt_txt'])
     d['temp'].append(i['main']['temp'])
     d['temp_min'].append(i['main']['temp_min'])
     d['temp_max'].append(i['main']['temp_max'])
     d['pressure'].append(i['main']['pressure'])
     d['humidity'].append(i['main']['humidity'])
     d['wind_speed'].append(i['wind']['speed'])
     d['clouad_all'].append(i['clouds']['all'])
     d['visibility'].append(i['visibility'])

d
df2= pd.DataFrame(d)
df2
```
